```python
import random
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

%precision 3
%matplotlib inline
```

```python
def N(mu, sigma, n=1):
    arr=[]
    for i in range(n):
        val = random.gauss(mu, sigma)
        arr.append(val)
    
    return  arr

N(90,10)
```

```python
def G(mu, sigma):
    def f(x):
        return (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x-mu)**2 / (2 * sigma**2))
    return  f
```

```python
mu, sigma = 0,4
sample = N(mu,sigma,1000)
dist = G(mu, sigma)
```

```python
x_min, x_max = -10,10
f = dist
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

xs = np.linspace(x_min, x_max, 100)
ax1.hist(sample, bins=100, range=(x_min, x_max), label='hist_dist')

ax2.plot(xs, [f(x) for x in xs], label='normal_dist', color='red')

ax1.legend()
ax2.legend()
plt.show()
```
